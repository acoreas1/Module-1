```python
import numpy as np
import pandas as pd
import os
import warnings
import re
from random import choice
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading csv file
df = pd.read_csv('/New Jersey.csv', on_bad_lines='skip')
display(df.head(5))
display(df.tail(5))
```

```python
# cleaning data
df.dropna(axis=1, how='all', inplace=True)
print(df.head(5))
print(df.tail(5))
```

```python
state_totals = {}
for col in df.columns:
    if col == 'date':
        continue
    total = 0
    for val in df[col]:
        if pd.notna(val):
            val_str = str(val).strip()
            is_negative = val_str.startswith('(') and val_str.endswith(')')
            val_clean = val_str.replace('$', '').replace(',', '').replace('(', '').replace(')', '')
            if val_clean.replace('.', '', 1).isdigit():
                val_num = float(val_clean)
                if is_negative:
                    val_num *= -1
                total += val_num
    state_totals[col] = total
print(state_totals)
```

```python
df2 = pd.DataFrame(list(state_totals.items()), columns=['State', 'Total Revenue'])
print(df2)
```
